### Encoding Categorical Variables

In this step, we apply several encoding techniques to prepare categorical variables for the model:

1. **One-Hot Encoding**: Used for categorical variables without any order (like 'CurrencyCode', 'ProviderId', 'ProductId', 'ProductCategory', and 'ChannelId'). This turns each category into a separate column with binary values.

2. **Label Encoding**: Applied to 'CountryCode', this method converts categories into numbers. It's useful for variables with some order.

3. **WOE (Weight of Evidence) Encoding**: A special technique used in credit scoring. It transforms categories (like 'CurrencyCode', 'ProviderId', 'ProductId', 'ProductCategory') based on their likelihood of default (whether a customer defaults on a loan or not). We also calculate **Information Value (IV)** to understand how predictive each feature is.

Errors during this process are handled and logged to ensure smooth execution.
